```python
df = df.drop("Unnamed: 0", axis = 1)
df
```

**GENDER DISTRIBUTION**

```python
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

*From the above chart we have Analysed that the number of females in the data is more than the number of males.*

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("relationship between parents education and student score")
plt.show()
```

*From the above chart we have concluded that the education of the parents have a good impact on their scores*

```python
gb_1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":'mean',"WritingScore":'mean'})
print(gb_1)
```

```python
sns.heatmap(gb_1, annot = True)
plt.title("relationship between parents marital status and student score")

plt.show()
```

*From the above chart we have Analysed that the marital status of the parents have no/negligible impact on student scores.*
