```python
import pandas as pd
import pickle
path = "/content/sample_data/knn_model.pickle"
with open(path,'rb') as f:
    knn_new = pickle.load(f)
```

```python

path_2 = "/content/sample_data/scaler.pickle"
with open(path_2,'rb') as f:
     scaler_new = pickle.load(f)
```

```python
path_3 = "/content/drive/MyDrive/Excel-and-CSV-files/new_customers.csv"
df = pd.read_csv(path_3)
df.head()
```

```python
len(df)
```

```python
gender_encoded_new = pd.get_dummies(df['Gender'],drop_first=True)
```

```python
gender_encoded_new
```

```python
df = pd.concat([df,gender_encoded_new],axis = 1)
```

```python
df
```

```python
x = df[['Male','Age','Salary','Price']].to_numpy()
x
```

```python
x_new_scaler = scaler_new.fit_transform(x)
x_new_scaler
```

```python
y_new_pred = knn_new.predict(x_new_scaler)
```

```python
y_new_pred
```

```python
df['will_purchase'] = y_new_pred
```
